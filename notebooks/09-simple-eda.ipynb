```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/processed/reddit_posts_2.parquet")
df.head(5)
df["search_query"].value_counts()
```

```python
# show heatmap where x axis is coin, y is subreddit and values are number of comments
df.groupby(['search_query', 'subreddit']).size().unstack().fillna(0).astype(int).plot(kind='bar', stacked=True, figsize=(20, 10))
```
